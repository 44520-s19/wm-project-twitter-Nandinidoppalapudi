### Installing tweepy and dumping the data from the chosen twitter accounts

```python
import tweepy
from twitter_keys import consumer_secret,access_token,consumer_key,access_secret
import pickle
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)

obama_tweets=api.user_timeline('BarackObama')
with open('obama_tweets1.pkl','wb')as f:
    pickle.dump(obama_tweets,f)

trump_tweets=api.user_timeline('realDonaldTrump')    
with open('Trump_tweets1.pkl','wb')as f:
    pickle.dump(trump_tweets,f)
    
ntr_tweets=api.user_timeline('worldNTRfans')    
with open('NTR_tweets1.pkl','wb')as f:
    pickle.dump(ntr_tweets,f)   
```

### Sentimental Analysis for the twitter accounts chosen

```python

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from visualization import get_sentiments
#build our figure
figure = plt.figure()
#build axes
ax=figure.add_subplot(111,projection='3d')

sentiments_analysis = [get_sentiments(tweet.text) for tweet in obama_tweets]

#build xs,ys,zs
xs=[]
for senti in sentiments_analysis:
    xs.append(senti['neg'])
    
ys=[senti['neu'] for senti in sentiments_analysis]

zs = list(map(lambda x:x['pos'],sentiments_analysis))

ax.scatter(xs,ys,zs)

#Another set of data - Donald Trump

    
from visualization import split_sentiments

a_X,a_Y,a_Z = split_sentiments([get_sentiments(tweet.text) for tweet in trump_tweets])
ax.scatter(a_X,a_Y,a_Z , color='g',marker='^')
ax.set_xlabel('Negative')
ax.set_ylabel('neutral')
ax.set_zlabel('postive')

#Another set of data - WorldNTRfans

from visualization import split_sentiments

b_X,b_Y,b_Z = split_sentiments([get_sentiments(tweet.text) for tweet in ntr_tweets])
ax.scatter(b_X,b_Y,b_Z, color='r',marker='s')
ax.set_xlabel('Negative')
ax.set_ylabel('neutral')
ax.set_zlabel('postive')

plt.show()
```

### Scatter plot for the obama and ntrfans accounts

```python
plt.scatter(a_X,a_Z, color='g',label='obama')
plt.scatter(b_X,b_Z, color='b',label='ntrfans')
plt.xlabel('Negative')
plt.ylabel('Positive')
plt.title('positive vs Negative sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```

### Average length of tweets for the accounts chosen

```python
from collections import Counter

print("Barack Obama : ")
sum=0
for tweet in obama_tweets:
    tweet1 = tweet.text.lower().split()
    print("Length of the tweet is ",len(tweet1))
    length = len(tweet1)
    sum=sum+length
print('sum is ',sum)
result=sum/len(obama_tweets)
print("\nAverage length tweets of- Barack Obama : ", result)
print()

print(" Donald Trump : ")
sum=0
for tweet in trump_tweets:
    tweet1 = tweet.text.lower().split()
    print("Length token is ",len(tweet1))
    length = len(tweet1)
    sum=sum+length
print('sum is ',sum)
result1=sum/len(trump_tweets)
print("\nAverage length  tweets of - Warren Buffet : ", result1)
print()

print("worldNTR Fans : ")
sum=0
for tweet in ntr_tweets:
    tweet1 = tweet.text.lower().split()
    print("Length token is ",len(tweet1))
    length = len(tweet1)
    sum=sum+length
print('sum is ',sum)
result2=sum/len(ntr_tweets)
print("\nAverage length tweets of - world NTR Fans : ", result2)
print()
```

### Average length of tweet represented using bar graph

```python
import matplotlib.pyplot as plt
heights = [result,result1,result2]
print(heights)
heights.sort()
print(heights)

left=[1,2,3]

tick_label=['Barack Obama', 'Donald Trump', 'World NTR Fans']

plt.bar(left,heights,tick_label=tick_label,color=['green'])

plt.xlabel('*Individual accounts*')

plt.ylabel('* Average *')
plt.title('Average length of tweets for an account')

plt.show()
```
